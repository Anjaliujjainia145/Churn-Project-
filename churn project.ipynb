```python
import matplotlib.pyplot as plt

# Data
labels = ['No', 'Yes', 'No internet service']
counts = [3473, 2044, 1526]
total = sum(counts)
percentages = [(count / total) * 100 for count in counts]
colors = ['#ff9999', '#66b3ff', '#ccccff']

# Plot
plt.figure(figsize=(4, 3))
bars = plt.bar(labels, percentages, color=colors, edgecolor='black')

# Add percentage labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 1,
             f'{height:.1f}%', ha='center', fontsize=12)

# Styling
plt.title('Tech Support Usage (Percentage)', fontsize=14)
plt.ylabel('Percentage (%)')
plt.ylim(0, max(percentages) + 10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['Not Senior Citizen', 'Senior Citizen']
churn_no = [4508, 666]
churn_yes = [1393, 476]

x = np.arange(len(categories))
width = 0.35

# Plot
fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, churn_no, width, label='No Churn', color='skyblue')
bars2 = ax.bar(x + width/2, churn_yes, width, label='Churn', color='salmon')

# Labels and Titles
ax.set_xlabel('Customer Type')
ax.set_ylabel('Number of Customers')
ax.set_title('Churn by Senior Citizen Status')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

# Show values on top of bars
for bar in bars1 + bars2:
    height = bar.get_height()
    ax.annotate(f'{height}', xy=(bar.get_x() + bar.get_width() / 1, height),
                xytext=(0, 3), textcoords="offset points", ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Prepare the data
data = {
    'Partner': ['No', 'No', 'Yes', 'Yes'],
    'Churn': ['No', 'Yes', 'No', 'Yes'],
    'Count': [2441, 1200, 2733, 669]
}

df = pd.DataFrame(data)

# Set style
sns.set(style="whitegrid")

# Plot
plt.figure(figsize=(4, 4))
scatter = sns.scatterplot(
    data=df,
    x='Partner',
    y='Churn',
    size='Count',
    sizes=(200, 1500),
    hue='Count',
    palette='coolwarm',
    legend='brief',
    alpha=0.7
)

# Customize
plt.title('Churn Status by Partner Status', fontsize=16)
plt.xlabel('Partner Status', fontsize=12)
plt.ylabel('Churn Status', fontsize=12)
plt.legend(title='Customer Count', bbox_to_anchor=(1.05, 1), loc='upper left')

# Annotate points
for i in range(df.shape[0]):
    plt.text(df['Partner'][i], df['Churn'][i], df['Count'][i],
             horizontalalignment='center', verticalalignment='center',
             fontsize=9, weight='bold', color='black')

plt.tight_layout()
plt.show()
```
